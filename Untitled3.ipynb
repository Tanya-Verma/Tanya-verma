**Hospital** emergency case arrival analysis minor 1 -statistics extension:Analyze delivery distribution across regions, Add-ons :visualization, stats,

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# -------------------------------------------------
# 1. DATA SIMULATION
# -------------------------------------------------
np.random.seed(2025)
regions = ['North-Metro', 'South-Rural', 'East-Industrial', 'West-Coastal']
emergency_types = ['Minor', 'Major', 'Critical']

data = {
    'Arrival_ID': range(1, 1001),
    'Region': np.random.choice(regions, 1000, p=[0.4, 0.15, 0.25, 0.2]),
    'Arrival_Time': pd.to_datetime('2025-01-01') + pd.to_timedelta(np.random.randint(0, 1440, 1000), unit='m'),
    'Emergency_Level': np.random.choice(emergency_types, 1000, p=[0.5, 0.3, 0.2]),
    'Response_Time_Min': np.random.gamma(shape=2, scale=7.5, size=1000)
}

df = pd.DataFrame(data)
df['Hour'] = df['Arrival_Time'].dt.hour

# -------------------------------------------------
# 2. STATISTICAL EXTENSION
# -------------------------------------------------
stats_summary = df.groupby('Region')['Response_Time_Min'].agg(['mean', 'median', 'std', 'count']).round(2)

contingency_table = pd.crosstab(df['Region'], df['Emergency_Level'])
chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print("--- 2025 Regional Statistical Summary ---")
print(stats_summary)
print(f"\nChi-Square Test p-value: {p_val:.4f}")

if p_val < 0.05:
    print("Result: Significant difference in case distribution across regions.")
else:
    print("Result: Case types are distributed uniformly across regions.")

# -------------------------------------------------
# 3. VISUALIZATION ADD-ONS
# -------------------------------------------------
sns.set_theme(style="whitegrid", palette="muted")
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

sns.countplot(data=df, x='Region', hue='Emergency_Level', ax=axes[0, 0])
axes[0, 0].set_title('Case Volume by Region and Severity (2025)', fontsize=14)

sns.violinplot(data=df, x='Region', y='Response_Time_Min', ax=axes[0, 1])
axes[0, 1].set_title('Response Time Density per Region', fontsize=14)

sns.lineplot(data=df, x='Hour', y='Arrival_ID', estimator='count', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Peak Emergency Arrival Hours', fontsize=14)
axes[1, 0].set_xticks(range(0, 24))

sns.heatmap(contingency_table, annot=True, fmt="d", cmap="YlGnBu", ax=axes[1, 1])
axes[1, 1].set_title('Region vs Emergency Type Heatmap', fontsize=14)

plt.tight_layout()
plt.show()

# -------------------------------------------------
# 4. EXPORTING RESULTS
# -------------------------------------------------
# df.to_csv('Hospital_Analysis_2025.csv', index=False)
```
